```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 6
sns.set_style('whitegrid')
```

### Eyeballing dataset distributions with histograms

```python
address= "D:/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/mtcars.csv"

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.index = cars.car_names
mpg = cars["mpg"]

mpg.plot(kind="hist")
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sns.displot(mpg)
```

###  Seeing scatterplots in action

```python
cars.plot(kind="scatter", x="hp", y="mpg", c=["darkgray"],s=150)
```

```python
sns.regplot(x="hp", y="mpg", data=cars, scatter=True)
```

### Generating a scatter plot matrix


```python
sns.pairplot(cars)
```

```python
cars_subset =  cars[["mpg","disp","hp","wt"]]
sns.pairplot(cars_subset)
plt.show()
```

### Building boxplots


```python
cars.boxplot(column="mpg", by="am")
cars.boxplot(column="wt", by="am")
```
