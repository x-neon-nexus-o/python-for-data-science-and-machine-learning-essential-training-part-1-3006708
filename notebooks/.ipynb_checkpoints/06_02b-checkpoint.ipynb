## Transforming Dataset Distributions

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, scale
```

```python
address = 'D:/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/mtcars.csv'

dataset = pd.read_csv(address)
dataset.head()
```

```python
plt.plot(dataset[["mpg"]])
```

### Normalization

```python
minmax_scalar=MinMaxScaler()
minmax_scalar.fit(dataset[["mpg"]])
```

```python
scaled_data=minmax_scalar.transform(dataset[["mpg"]])
plt.plot(scaled_data)
```

### Standardization

```python
standard_scalar=scale(dataset[["mpg"]])
plt.plot(standard_scalar)
```
